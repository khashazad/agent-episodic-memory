```python
from textwrap import wrap
import json

def display_collection_documents(collection, num_docs=3):
    """Display documents from a collection in a human-friendly format."""
    print("=" * 80)
    print(f"Collection: {collection.name}")
    print(f"Total documents: {collection.count()}")
    print("=" * 80)
    print()

    # Get first few documents
    results = collection.get(limit=num_docs, include=["documents", "metadatas"])

    for i, (doc_id, doc, metadata) in enumerate(zip(
        results.get("ids", []),
        results.get("documents", []),
        results.get("metadatas", [None] * len(results.get("ids", [])))
    ), 1):
        print(f"\n--- Document {i} (ID: {doc_id}) ---")

        # Always display metadata section
        print("\nMetadata:")
        if metadata:
            for key, value in sorted(metadata.items()):
                if value is None or value == "":
                    print(f"  {key}: (empty)")
                elif isinstance(value, (dict, list)):
                    print(f"  {key}:")
                    print(f"    {json.dumps(value, indent=6)}")
                else:
                    print(f"  {key}: {value}")
        else:
            print("  (no metadata)")

        # Display document text with word wrapping
        print("\nDocument Text:")
        print("-" * 80)
        # Wrap text to 80 characters for readability
        wrapped_text = wrap(doc, width=80) if doc else ["(empty)"]
        for line in wrapped_text:
            print(line)
        print("-" * 80)
        print()

# Display documents from both collections
collections_to_view = ["episodic_memory_v1", "episodic_memory_v2"]

for collection_name in collections_to_view:
    try:
        collection = client.get_collection(collection_name)
        display_collection_documents(collection, num_docs=3)
        print("\n" * 2)
    except Exception as e:
        print(f"Error accessing collection '{collection_name}': {e}")
        print()
```
